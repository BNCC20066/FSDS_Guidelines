# Custom Loss Function

## Binary Cross Entropy Loss

Initialize value

```python
y_true=np.random.randint(0,2,20)
y_pred = np.random.uniform(size=20)
```

```python
y_pred
```

Keras Function

```python
bce = keras.losses.BinaryCrossentropy(from_logits=False)
bce(y_true, y_pred)
```

Binary Cross Entropy loss is defined by
\begin{equation}
L=-(y.log(\hat{y})+(1-y).log(1-\hat{y}))
\end{equation}

```python
def bce_fn(y_true, y_pred):
  epsilon = 1e-07
  error = y_true*math_ops.log(y_pred)
  error += (1-y_true+epsilon)*math_ops.log(1-y_pred+epsilon)
  error = -1*tf.math.reduce_mean(error)
  return error
```

```python
bce_fn(y_true,y_pred)
```
